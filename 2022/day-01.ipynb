```python
with open('input-01.txt') as f:
    lines = f.read()
lines
```

```python
data = lines[:-1]
data = data.split('\n\n')
data = [item.split('\n') for item in data]
data = [[int(item) for item in sublist] for sublist in data]
data
```

```python
data_sums = [sum(sublist) for sublist in data]
data_sums
```

```python
answer1 = max(data_sums)
answer1
```

```python
largest_ones = data_sums
largest_ones.sort(reverse=True)
answer2 = sum(largest_ones[:3])
answer2
```
