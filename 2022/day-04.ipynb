```python
with open('input-04.txt') as f:
    lines = f.read()
lines
```

```python
data = lines
data = data.split('\n')[:-1]
data[:8]
```

```python
sections = [couple.split(',') for couple in data]
sections = [[section.split('-') for section in couple] for couple in sections]
sections = [[[int(number) for number in section] for section in couple] for couple in sections]
sections[:8]
```

```python
ranges = [[set(range(item[0],item[1]+1)) if item[0] != item[1] else set([item[0]]) for item in couple] for couple in sections]
ranges[:-8]
```

```python
print(ranges[:8])
```

```python
contained = [setA.issubset(setB) or setB.issubset(setA) for setA, setB in ranges]
contained
```

```python
answer1 = sum(contained)
answer1
```

```python
overlap = [not setA.isdisjoint(setB) for setA, setB in ranges]
overlap
```

```python
answer2 = sum(overlap)
answer2
```
