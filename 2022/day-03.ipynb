```python
with open('input-03.txt') as f:
    lines = f.read()
lines
```

```python
data = lines
data = data.split('\n')[:-1]
data
```

```python
common = [[character for character in word if (character in word[:int(len(word)/2)] and character in word[int(len(word)/2):])] for word in data]
common
```

```python
def checker(word):
    word_0 = word[:int(len(word)/2)]
    word_1 = word[int(len(word)/2):]
    return [character for character in word if (character in word_0 and character in word_1)]

common = [checker(word)[0] for word in data]
common
```

```python
characters = [character for character in 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ']
priority = range(1,len(characters)+1)
evaluate = dict(zip(characters, priority))
evaluate
```

```python
priorities = [evaluate[character] for character in common]
priorities
```

```python
answer1 = sum(priorities)
answer1
```
