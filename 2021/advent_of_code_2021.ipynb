```python
# handy function to get a whole column of bits

def nth_column(list_of_strings, n):
  return ["".join([row[position] for row in list_of_strings]) for position in range(len(list_of_strings[0]))][n]
```

```python
from itertools import *
```

```python
def rate(list_of_strings, default_value):

  # if there's a single item in the list, job is already done
  if len(list_of_strings) == 1:
    return list_of_strings[0]

  # otherwise, looking for the first column with non identical bits
  position = 0
  column = nth_column(list_of_strings, position)
  while len(set(column)) != 2:
    position += 1
    column = nth_column(list_of_strings, position)

  # getting a reference value according with the type of rating (O2/CO2)
  if column.count('0') == column.count('1'):
    value = str(default_value)
  elif default_value == 1:
    value = max(set(column), key=column.count)
  else:
    value = min(set(column), key=column.count)

  # filtering the list according with that reference value
  new_list = list(compress(list_of_strings, [row[position] == value for row in list_of_strings]))

  # feeding the filtered to the same function, recursion will do the rest
  return rate(new_list, default_value)
```

```python
class BoardBingo():

  def __init__(self, board):
    self.board = board
    self.board_flat = [item for row in self.board for item in row]
    self.status_flat = [0 for i in range(25)]
  
  def __repr__(self):
    return " ".join([str(item) for item in self.board_flat])

  def check_number(self, number):
    if number in self.board_flat:
      self.status_flat[self.board_flat.index(number)] = 1

  def check_win(self):
  
    # by row
    self.status_rows = [self.status_flat[i:i + 5] for i in range(0, 25, 5)]
    test_row =  5 in [sum(chunk) for chunk in self.status_rows]
    
    # by column
    self.status_column = [self.status_flat[i::5] for i in range(5)]
    test_column =  5 in [sum(chunk) for chunk in self.status_column]

    return test_row or test_column

boards_BB = [BoardBingo(board) for board in boards]
boards_BB[:8]
```

```python
winning_board = None
tot_num_drawned = None

try: 
  for number in numbers:
    for board in boards_BB:
      board.check_number(number)
      if board.check_win():
        winning_board = boards_BB.index(board)
        tot_num_drawned = numbers.index(number) + 1 
        print(f"Winner is board {winning_board} after drawning {tot_num_drawned} numbers")
        raise StopIteration
except StopIteration: pass
```
